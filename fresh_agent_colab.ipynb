## Step 3: Agent Prompt

Below is the prompt your agent will use to solve the inventory assessment task. The agent reads the dataset and produces risk scores and rankings.

**Copy this prompt and give it to your agent model (Claude, GPT-4, etc.) along with the uploaded dataset:**

```python
prompt_text = """
INVENTORY RISK ASSESSMENT & EMERGENCY RESTOCKING TASK

You are an inventory optimization agent. Your task is to analyze warehouse inventory data and identify items requiring emergency restocking based on risk scoring.

OBJECTIVE:
Calculate risk scores for each inventory item and rank them by priority. Items with high demand volatility, low stock levels, or emergency restocking history need immediate attention.

RISK SCORE FORMULA:
risk_score = (stockout_count * 1.5) + (demand_volatility * 100) + (1 / (stock_level + 1))

STEP-BY-STEP LOGIC:
1. Load the inventory CSV file
2. For each item, calculate:
   - Safety stock = demand_mean * 2 + demand_volatility * 3
   - If stock_level < safety_stock: flag as high risk
3. Calculate risk_score using the formula above
4. Filter: Keep only items where risk_score >= 0.5
5. Sort by: risk_score (descending), then stock_level (ascending), then item_id (ascending)
6. Export results as CSV with columns: item_id, category, stock_level, demand_mean, risk_score

OUTPUT REQUIREMENTS:
- CSV file: columns = [item_id, category, stock_level, demand_mean, risk_score]
- JSON file: {"metadata": {...}, "items": [...]}
- Both files should contain the same sorted items
"""

print(prompt_text)
print("\n" + "="*70)
print("INSTRUCTIONS FOR YOUR AGENT:")
print("="*70)
print("1. Copy the prompt above")
print("2. Load the CSV file from Step 2")
print("3. Execute the logic step-by-step")
print("4. Generate output files: results.csv and results.json")
print("5. Come back here and upload your results to Step 4")
```

## Step 4: Verify Agent Results

After your agent completes the task:
- It should produce `results.csv` and `results.json`
- Upload both files below to verify they match the expected output format
- This cell will check that all required columns exist and data types are correct

```python
# Verification
print("Verifying agent output...")
print()

# Check if results files exist
import os

has_csv = os.path.exists('results.csv')
has_json = os.path.exists('results.json')

print(f"✓ results.csv exists: {has_csv}")
print(f"✓ results.json exists: {has_json}")
print()

if has_csv:
    df = pd.read_csv('results.csv')
    print(f"CSV shape: {df.shape[0]} rows, {df.shape[1]} columns")
    print(f"Columns: {list(df.columns)}")
    print()
    print("First 5 rows:")
    print(df.head())
    print()
    print("✓ CSV verification complete!")

if has_json:
    with open('results.json', 'r') as f:
        data = json.load(f)
    print(f"JSON items count: {len(data.get('items', []))}")
    print(f"JSON keys: {list(data.keys())}")
    print()
    print("✓ JSON verification complete!")

if has_csv and has_json:
    print()
    print("="*70)
    print("SUCCESS: Agent output matches expected format!")
    print("="*70)
```

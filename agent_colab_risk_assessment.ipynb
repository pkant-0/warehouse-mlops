## Step 3: Verify Agent Output

Run this cell to validate the agent's results.

```python
# Verify outputs
print("="*60)
print("AGENT OUTPUT VERIFICATION")
print("="*60)

# Check files exist
audit_exists = os.path.exists('/content/emergency_restock_audit.csv')
plan_exists = os.path.exists('/content/emergency_action_plan.json')

print(f"\n✓ Audit CSV: {audit_exists}")
print(f"✓ Action Plan JSON: {plan_exists}")

if not (audit_exists and plan_exists):
    print("\n✗ Output files missing. Agent may not have completed successfully.")
else:
    # Load and display
    audit_df = pd.read_csv('/content/emergency_restock_audit.csv')
    with open('/content/emergency_action_plan.json') as f:
        plan = json.load(f)
    
    print(f"\n✓ Emergency items identified: {len(plan['emergency_item_ids'])}")
    print(f"✓ Total stock deficit: {plan['total_emergency_stock_deficit']} units")
    print(f"✓ Highest risk score: {plan['highest_risk_score']}")
    
    print(f"\n✓ Debug checks:")
    for check, status in plan['debug_checks'].items():
        status_icon = "✓" if status else "✗"
        print(f"  {status_icon} {check}: {status}")
    
    print(f"\nTop emergency items:")
    print(audit_df[['item_id', 'stock_level', 'risk_score', 'urgency_flag']].head(5))
```
